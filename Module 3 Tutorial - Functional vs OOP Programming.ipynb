```python
# Sorting 0s, 1s, 2s + Binary Search Assignment
```

```python
class Solution:
    def sort012(self, arr):
        """
        Sorts an array containing only 0s, 1s, and 2s
        using Dutch National Flag algorithm.
        Time: O(n), Space: O(1)
        """
        low = mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1

            elif arr[mid] == 1:
                mid += 1

            else:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr
```

```python
obj = Solution()

print(obj.sort012([0, 1, 2, 0, 1, 2]))
print(obj.sort012([0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]))
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1

        while low <= high:
            mid = (low + high) // 2

            if arr[mid] == k:
                result = mid
                high = mid - 1  # search left for first occurrence

            elif arr[mid] < k:
                low = mid + 1

            else:
                high = mid - 1

        return result
```

```python
obj = Solution()

print(obj.binarysearch([1, 2, 3, 4, 5], 4))
print(obj.binarysearch([11, 22, 33, 44, 55], 445))
print(obj.binarysearch([1, 1, 1, 1, 2], 1))
```
